# Machine Learning（大话版）
> 在开始了解机器学习的概念之前，有一些前置的基础知识，如线性代数、概率论等，关于为什么要进行机器学习、机器学习能做什么之类的“废话”，统统不会在下面出现。此note将介绍机器学习的基本概念与流程，及其相应的评价指标，在最后将手写一个分类算法、一个回归算法、一个无监督算法，进行“感性”地认识。并且该note，采用比较白话的叙述方式，尽可能用通俗易懂的语言定义和描述其中出现的知识。
## 机器学习必须要知道的概念
1. **有监督学习和无监督学习**：如果刚看到这个概念很蒙，我们不妨把上面两个词进行拆分理解。
   1. 首先，**“学习”**是一个什么样的过程？通俗地讲，一个人学习加法运算的过程，就是从不知道如何对两个数进行加法运算到会对任意数进行加法运算的过程。在学习加法运算的过程中，这个人获得的加法运算的经验，就是知识，这个过程就是学习。那么，**无论是有监督学习还是无监督学习都是获取知识的过程。**
   2. 同样都是获取知识，有监督和无监督的学习有什么差别呢？**有监督学习是一种带有“标签”的学习**，也就是说能知道对错的学习，就拿上文中的学习加法的过程，可以知道1+1=2，这是正确还是错误的。在学习加法运算的过程中，可以根据加法运算结果的正确与错误进行调整，从何学地更好更快。**那无监督学习就是没有标签的学习**，比如将一车西瓜的大小划分成小、中、大，在没有规定大于多少g的规定下进行划分，这个划分的过程就是无监督学习的过程，听着很玄乎，但是早有人针对各种问题提出了不同的方法。
2. **分类和回归**
   1. 分类：对一个物体（样本）按其某一属性进行类别的划分，对男生和女生进行性别的划分。
   2. 回归： 回归其实很简单，最常见的就是，初高中学的物理题，已知，物体A在做加速度不变（大小和方向都不变）的直线运动，物体A分别的t1、t2...tn时刻的速度为v1、v2...vn，计算tx时刻，物体A的速度。这个种类型的题利用牛顿三定律非常容易求解，但是如果我们不知道牛顿定律，我们该怎么办呢？如果给我们足够的点，我们是不是可以把连起来，用一个“连起来的函数”来估计物体A在tx时刻的速度，通过足够的点来估计这个函数的任务就叫回归。
> 那么分类和回归是有监督学习还是无监督学习呢？
3. **样本与数据集**
   在机器学习中样本是一个抽象的概念，它可以描述一个人，也可以描述一个物体，比如说，在男女分类时，一个人就是一个样本，在西瓜分大小时，一个西瓜就是一个样本，许多样本可以构成数据集。也就是说 数据集是许多样本的集合。
3. **过拟合和欠拟合**
  1. 过拟合：就是学习的过程中由于“样本”不均衡等原因导致，学到的知识带有“刻板印象”，比如说男女性别分类，在学习的过程中，看到穿裙子的都是女孩，学到的经验就是穿裙子的一定是女孩，可是还有“女装大佬”，男孩子也可以穿裙子呀，学到的这个经验是不对的，会影响后面的判断。如果没有学习到还有女装大佬这个知识，考试的时候，给出穿裙子的男孩，根据以前的知识，判断出是女孩，这个过程就是过拟合。
  2.掐拟合：这个就更好理解了，就是学习的知识不到位。还是男女性别分类这个例子，有长胡子的一定是男生，这知识没有学习到，这就是欠拟合。
